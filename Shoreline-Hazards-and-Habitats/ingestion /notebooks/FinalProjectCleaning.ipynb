My code below merges my two datasets using OBJECTID so all information is combined into one dataframe called merged. It then cleans and standardizes the data by converting habitat presence columns from “Yes/No” into numeric values "0/1". Next, it converts several hazard-related columns from categorical labels like “Low” or “High” into numeric scores 0–4 so they can be used in statistical analysis. Finally, I made sure the main hazard variables were treated as ordered categories.

```python
#used built in Gemini

habitats = [
    "Forest_Present",
    "Marsh_Present",
    "Dune_Present",
    "Oyster_Reef_Present",
    "Underwater_Grass_Present"
]
hazard_mapping = {
    'Very Low': 0,
    'Low': 1,
    'Moderate': 2,
    'High': 3,
    'Very High': 4
}
hazard_order = ["Very Low", "Low", "Moderate", "High", "Very High"]
hazard_cols = [
    'Geomorphology_Hazard', 'Elevation_Hazard',
    'Sea_Level_Rise_Hazard', 'Wave_Hazard',
    'storm_surge_hazard', 'Erosion_Rate_Hazard'
]


merged = pd.merge(md, marsh_data, on='OBJECTID', how='left')


for h in habitats:
    if h in merged.columns:

        merged.loc[:, h] = merged[h].map({"Yes": 1, "No": 0}).fillna(0).astype(int)


if 'Hazard_Reduction_by_Habitats' in merged.columns:
    merged.loc[:, 'Hazard_Reduction_by_Habitats_Numeric'] = merged['Hazard_Reduction_by_Habitats'].map(hazard_mapping)
if 'Hazard_with_Habitats' in merged.columns:
    merged.loc[:, 'Hazard_with_Habitats_Numeric'] = merged['Hazard_with_Habitats'].map(hazard_mapping)
if 'Hazard_without_Habitats' in merged.columns:
    merged.loc[:, 'Hazard_without_Habitats_Numeric'] = merged['Hazard_without_Habitats'].map(hazard_mapping)


if 'Community_Flood_Risk_Max' in merged.columns:
    merged.loc[:, 'Community_Flood_Risk_Max_Numeric'] = merged['Community_Flood_Risk_Max'].map(hazard_mapping)


for col in hazard_cols:
    if col in merged.columns:
        merged.loc[:, col] = pd.Categorical(merged[col], categories=hazard_order, ordered=True)
```

I saw that my Shoreline_Priority_Area variable had many missing values. If I removed all the rows where this variable was missing, I would lose close to half of my data. Because of this, I decided to remove the column. My other variables that has missing values also had pretty large amounts of missing data, but not enough that I wouldn't resort to cutting them out completely.

```python
#Used built in Gemini

print(f"Merged Data Original Shape {merged.shape}")


if 'Shoreline_Priority_Area' in merged.columns:
    merged = merged.drop(columns=['Shoreline_Priority_Area'])
    print(f"Shape after dropping 'Shoreline_Priority_Area': {merged.shape}")
else:
    print("'Shoreline_Priority_Area' column not found.")


remaining_nan_cols = merged.columns[merged.isnull().any()].tolist()

if remaining_nan_cols:
    print(f"Remaining columns with NaN values: {remaining_nan_cols}")
    print(f"Shape before dropping rows with NaNs in specific columns: {merged.shape}")
    merged = merged.dropna(subset=remaining_nan_cols)
    print(f"Shape after dropping rows with NaNs in specific columns: {merged.shape}")
else:
    print("No remaining columns with NaN values after dropping 'Shoreline_Priority_Area'.")


display(merged.head())


print("\nMissing values after cleanup:")
display(merged.isnull().sum()[merged.isnull().sum() > 0])
```
